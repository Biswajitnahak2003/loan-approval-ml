```python
# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\biswa\classification_problem\loan_approval_pred\comparison_of_all\comparison.csv")
df = df.sort_values(by='f1_score', ascending=False) 

sns.set(style="whitegrid")
plt.figure(figsize=(14, 6))

plt.subplot(1, 3, 1)
sns.barplot(data=df, x='model_name', y='f1_score', palette='Blues_d')
plt.title('F1 Score Comparison')
plt.xticks(rotation=45)

plt.subplot(1, 3, 2)
sns.barplot(data=df, x='model_name', y='accuracy', palette='Greens_d')
plt.title('Accuracy Comparison')
plt.xticks(rotation=45)

plt.subplot(1, 3, 3)
sns.barplot(data=df, x='model_name', y='roc_auc', palette='Oranges_d')
plt.title('ROC-AUC Comparison')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
